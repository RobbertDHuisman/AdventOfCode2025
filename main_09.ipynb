```python
from pathlib import Path
from dataloader.dataloader import Dataloader
from day_09.movie_theater import MovieTheater
import numpy as np
```

## Load and Parse Data

```python
data = Dataloader(Path("day_09/input.csv")).data

red_tiles = []
for line in data:
    coordinates = []
    point = line.split(',')
    for coordinate in point:
        coordinates.append(int(coordinate))
    red_tiles.append(coordinates)

red_tiles = np.array(red_tiles)
```

## Create Movie Theater and Calculate Areas

```python
movie_theater = MovieTheater(red_tiles)
```

```python
movie_theater.tile_matrix[movie_theater.tile_matrix == 2] = 1
```

```python
x0, y0 = movie_theater.red_tiles[1]
x1, y1 = movie_theater.red_tiles[3]
matrix = movie_theater.tile_matrix[min(y0, y1):max(y0, y1)+1, min(x0, x1):max(x0, x1)+1]
print(matrix)
```

```python
new_max_area = 0
new_area_matrix = movie_theater.area_matrix.copy()
while new_max_area == 0:
# for i in range(2):
    max_idx = np.unravel_index(new_area_matrix.argmax(), new_area_matrix.shape)
    print("Max idx:", max_idx)

    first, second = max_idx
    x0, y0 = movie_theater.red_tiles[first]
    x1, y1 = movie_theater.red_tiles[second]

    if (movie_theater.tile_matrix[y0, x1] != 0) and (movie_theater.tile_matrix[y1, x0] != 0):
        matrix = movie_theater.tile_matrix[min(y0, y1):max(y0, y1)+1, min(x0, x1):max(x0, x1)+1]
        if np.where(matrix == 0)[0].size == 0:
            new_max_area = np.sum(matrix)

    new_area_matrix[(first,second)] = 0
    new_area_matrix[(second, first)] = 0



print("Max area:", new_max_area)
            
```

## Results

```python
print(f"Part 1: area = {movie_theater.area_matrix.max()}")
print(f"Part 2: new area = {new_max_area.max()}")
```
